# Images Reading - Plotting

### Read Image

```python
import skimage.io as io
import numpy as np
imagen = io.imread(fname="images/imageonline-co-blackandwhiteimage.jpg",as_gray=True)
#imagen= imagen_ori[:,:,0]
display(imagen)
display(imagen.shape)
display(type(imagen))
display(np.max(imagen))
display(np.min(imagen))
display(np.mean(imagen))
display(np.median(imagen))
display(np.std(imagen))
```

### Alter Image

```python
import matplotlib.pyplot as plt
imagen_nueva = imagen.copy()
display(len(imagen_nueva[imagen_nueva>=0.95]))
imagen_nueva[imagen_nueva>=0.70]=0
fig,ax=plt.subplots(1,2,figsize=[7,7])
ax[0].imshow(imagen_nueva,cmap=plt.cm.gray)
ax[1].imshow(imagen,cmap=plt.cm.gray)
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,4,figsize=[20,4])
ax[0].hist(imagen.ravel(),bins=255)
ax[1].hist(imagen.ravel(),bins=100)
ax[2].hist(imagen.ravel(),bins=25)
ax[3].hist(imagen.ravel(),bins=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
imagen_nueva_contraste=imagen.copy()
imagen_nueva_contraste[(imagen_nueva_contraste>0.62) &(imagen_nueva_contraste<0.75 )]=0
imagen_nueva_contraste[(imagen_nueva_contraste>=0.01) &(imagen_nueva_contraste<0.175 )]=1
fig,ax=plt.subplots(1,2,figsize=[10,5])
ax[0].imshow(imagen_nueva_contraste,cmap=plt.cm.gray)
ax[1].imshow(imagen,cmap=plt.cm.gray)
plt.show()
fig,ax=plt.subplots(2,4,figsize=[20,4])
ax[0,0].hist(imagen.ravel(),bins=255)
ax[0,1].hist(imagen.ravel(),bins=100)
ax[0,2].hist(imagen.ravel(),bins=25)
ax[0,3].hist(imagen.ravel(),bins=10)
ax[1,0].hist(imagen_nueva_contraste.ravel(),bins=255)
ax[1,1].hist(imagen_nueva_contraste.ravel(),bins=100)
ax[1,2].hist(imagen_nueva_contraste.ravel(),bins=25)
ax[1,3].hist(imagen_nueva_contraste.ravel(),bins=10)
plt.show()
```
