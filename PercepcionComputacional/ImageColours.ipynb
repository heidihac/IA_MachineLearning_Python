### Image Colours

```python
import skimage.io as io
imagen = io.imread(fname="images/IMG_3864.JPG")
```

```python
import matplotlib.pyplot as plt
def mostrar_imagen(imagen,grayscale=False):
    fig,ax=plt.subplots(figsize=[7,7])
    if grayscale:
        ax.imshow(imagen,cmap='gray')        
    else:
        ax.imshow(imagen)
    plt.show()
```

```python
mostrar_imagen(imagen)
```

```python
import numpy as np
total=imagen.ravel()
red=imagen[:,:,0]
green=imagen[:,:,1]
blue=imagen[:,:,2]
plt.hist(total,bins=256,color='orange',alpha=0.5,)
plt.hist(red.ravel(),bins=256,color='red',alpha=0.5)
plt.hist(green.ravel(),bins=256,color='green',alpha=0.5)
plt.hist(blue.ravel(),bins=256,color='blue',alpha=0.5)
plt.xlim([0,256])
plt.show()
```

```python
img=np.zeros(imagen.shape, dtype=int)
img[:,:,0]=red
mostrar_imagen(img)
img=np.zeros(imagen.shape, dtype=int)
img[:,:,1]=green
mostrar_imagen(img)
img=np.zeros(imagen.shape, dtype=int)
img[:,:,2]=blue
mostrar_imagen(img)
```

```python
from skimage import img_as_float
imagen_float=img_as_float(imagen)
mostrar_imagen(imagen_float)
```

```python
mostrar_imagen(red,True)
mostrar_imagen(green,True)
mostrar_imagen(blue,True)
```

```python
img_mix=np.zeros(imagen.shape)
img_mix[:,:,0]=imagen_float[:,:,2]
img_mix[:,:,1]=imagen_float[:,:,1]
img_mix[:,:,2]=imagen_float[:,:,0]
mostrar_imagen(img_mix)
img_mix=np.zeros(imagen.shape)
img_mix[:,:,0]=imagen_float[:,:,2]
img_mix[:,:,1]=imagen_float[:,:,0]
img_mix[:,:,2]=imagen_float[:,:,1]
mostrar_imagen(img_mix)
img_mix[:,:,0]=imagen_float[:,:,1]
img_mix[:,:,1]=imagen_float[:,:,2]
img_mix[:,:,2]=imagen_float[:,:,0]
mostrar_imagen(img_mix)
```
