# No Lineal Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Grafica Regrsion Lineal

```python
x=np.arange(-5.0,5.0,0.1)
y=2*(x)+3
plt.show()
y_noise=2*np.random.normal(size=x.size)
ydata=y+y_noise
plt.plot(x,ydata,'ob')
plt.plot(x,y,'r')
plt.xlabel('Var Independiente')
plt.ylabel('Var Dependiente')
plt.show()
```

### Grafica Regresion No Lineal Cubica

```python
x = np.arange(-5,5,0.1)
y = x**3+x**2+x**1+1
y_noise = 20*np.random.normal(size=x.size)
ydata = y+y_noise
plt.plot(x,ydata,'ob')
plt.plot(x,y,'r')
plt.xlabel('Var Independiente')
plt.ylabel('Var Dependiente')
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y=x**2
y_noise=6*np.random.normal(size=x.size)
ydata=y+y_noise
plt.plot(x,ydata,'ob')
plt.plot(x,y,'r')
plt.xlabel('Var Independiente')
plt.ylabel('Var Dependiente')
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y=10+5*(2**x)
y_noise=10*np.random.normal(size=x.size)
ydata=y+y_noise
plt.plot(x,ydata,'ob')
plt.plot(x,y,'r')
plt.xlabel('Var Independiente')
plt.ylabel('Var Dependiente')
plt.show()
```

```python
x=np.arange(0.1,5,0.1)
y=15-2*np.log(x)
y_noise=2*np.random.normal(size=x.size)
ydata=y+y_noise
plt.plot(x,ydata,'ob')
plt.plot(x,y,'-r')
plt.xlabel('Var Independiente')
plt.ylabel('Var Dependiente')
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y=1-4/(1+np.power(3,x-2))
y_noise=np.random.normal(size=x.size)/4
ydata=y+y_noise
plt.plot(x,ydata,'ob')
plt.plot(x,y,'-r')
plt.xlabel('Var Independiente')
plt.ylabel('Var Dependiente')
plt.show()
```
