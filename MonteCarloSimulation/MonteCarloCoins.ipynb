## Monte Carlo

### Dice

```python
import random
random.seed(1234)
def rollDice():
    roll = random.randint(1,100)
    if roll == 100:
        #print('You lose')
        return False
    elif roll <=50:
        #print('Rolla 1-50, you lose')
        return False
    elif 100>roll>50:
        #print('You Win')
        return True
def simple_bettor(funds,initial_wager,wager_count):
    value = funds
    wager = initial_wager
    wX=[]
    vY=[]
    currentWager = 1
    while currentWager<=wager_count:
        if rollDice():            
            value += wager            
        else:
            value -= wager
        wX.append(currentWager)
        vY.append(value)
        currentWager+=1
    #if value < 0:
        #print('broke')
    #print('Founds',value)
    return wX,vY
```

```python
counter =0 
while counter<100:
    simple_bettor(10000,100,5000)
    counter +=1
```

```python
import matplotlib.pyplot as plt
counter =0 
plt.figure(figsize=(20,10))
while counter<100:
    wX,vY=simple_bettor(10000,100,1000)
    plt.plot(wX,vY)
    counter +=1

plt.ylabel('Value')
plt.xlabel('Wawer')
```
