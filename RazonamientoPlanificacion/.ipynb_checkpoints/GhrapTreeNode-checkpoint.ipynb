# Tree Node Graph

```python
import networkx as nx
G = nx.DiGraph()
G.add_node(1, time='5pm')
G.add_node(2, time='2pm')
G.add_node(3, time='1pm')
```

```python
G.add_edge(1,2)
G.add_edge(1,3)
```

```python
import matplotlib.pyplot as plt
nx.draw(G, with_labels = True)
```

```python
import networkx as nx

G = nx.DiGraph()

G.add_node("ROOT")

for i in range(5):
    G.add_node("Child_%i" % i)
    G.add_node("Grandchild_%i" % i)
    G.add_node("Greatgrandchild_%i" % i)

    G.add_edge("ROOT", "Child_%i" % i)
    G.add_edge("Child_%i" % i, "Grandchild_%i" % i)
    G.add_edge("Grandchild_%i" % i, "Greatgrandchild_%i" % i)


# same layout using matplotlib with no labels
plt.title('draw_networkx')
nx.draw(G,with_labels = True)
```

```python
import networkx as nx
class Nodo:
    def __init__(self,nodo,nodosVecinos=None):
        self.nodo=nodo
    def __str__(self):
        return self.nodo
A = Nodo("A")
B = Nodo("B")
C = Nodo("C")
D = Nodo("D")
E = Nodo("E")
F = Nodo("F")
G = Nodo("G")
A.nodosVecinos = [A,B,C]
B.nodosVecinos = [A,B,D,E]
C.nodosVecinos = [A,F]
D.nodosVecinos = [B,G]
E.nodosVecinos = [B,F,G]
F.nodosVecinos = [C,E]
G.nodosVecinos = [D,E]
    
grafico={
    A:A.nodosVecinos,
    B:B.nodosVecinos,
    C:C.nodosVecinos,
    D:D.nodosVecinos,
    E:E.nodosVecinos,
    F:F.nodosVecinos,
    G:G.nodosVecinos    
}
G=nx.DiGraph(grafico)
nx.draw_spectral(G,with_labels=True, font_weight='bold')
```

```python
figura = {}
nodo=A
nodoVisitado=[]
def crear_diccinario(nodo):
    if not nodo:
        return -1
    else:        
        if not nodo in nodoVisitado:
            figura[nodo] = nodo.nodosVecinos
            nodoVisitado.append(nodo)
            for nodoVecino in nodo.nodosVecinos:
                crear_diccinario(nodoVecino)
crear_diccinario(A)
G=nx.DiGraph(figura)
nx.draw_spectral(G,with_labels=True, font_weight='bold') 
```
