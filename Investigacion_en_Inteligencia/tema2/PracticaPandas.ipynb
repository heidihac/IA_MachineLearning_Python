```python
import random
df=pd.DataFrame({'A1':[1],'A2':['Luis']})
display(df)
dates=pd.date_range('01/01/2020',periods=4,freq='M')
df=pd.DataFrame({'C1':pd.Timestamp('30-04-1986'),'C2':[1,2,3,4],
                'C3':['A','B','C','D'],
                'C4':np.asanyarray([33]*4,dtype="float"),
                'C5':np.asanyarray([random.randint(1,100)]*4,dtype="float"),
                'C6':pd.Categorical(['A',2,'C',2]),
                'C7':'Cumple',
                'C8':['Ab','Bc','Ca','Db']
                },index=dates)
display(df)
display(df.index)
display(df.index.dtype)
display(df['C6'].dtype)
print(df.values)
display(df.describe())
```
