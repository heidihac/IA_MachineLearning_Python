```python
import pandas as pd
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3'],
                    'E': [1,2,3,4]},
                   index=[0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])

df5 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11'],
                    'F': [110, 120, 'D10', 'D11']},
                   index=['a', 'e','i', 'o'])
frame = [df1,df2,df3]
result=pd.concat(frame)
display(result)
display(pd.concat(frame,axis=1))
display(pd.concat(frame,join="inner"))
display(pd.concat([df1,df5],join="outer",ignore_index=True))
display(pd.concat([df1,df2,df5],ignore_index=False))
df_xy=pd.concat([df1,df3],keys=['x','y'],names =['group','indexes'])
display(df_xy)
display(df_xy.loc['x',['A','B','C']])
```

```python
display(pd.concat([df1,df4],axis=1,sort=False))
display(pd.concat([df1,df4],axis=1,join="inner"))
display(df1.index)
display(df4.index)
display(df4.reindex(df1.index))
display(pd.concat([df1,df4.reindex(df1.index)],axis=1))
```

# MERGE

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
display(pd.merge(left,right))
display(pd.merge(left,right,on="key"))

right = pd.DataFrame({'key': ['K2', 'K3', 'K4', 'K5'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
display(pd.merge(left,right,on="key"))
display(pd.merge(left,right,on="key",how="inner"))
display(pd.merge(left,right,on="key",how="left"))
display(pd.merge(left,right,on="key",how="right"))
display(pd.merge(left,right,on="key",how="outer"))
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
display(pd.merge(left,right,on=["key1"]))
display(pd.merge(left,right,on=["key1","key2"]))
display(pd.merge(left,right,on=["key1","key2"],how="left"))
display(pd.merge(left,right,on=["key1","key2"],how="right"))
display(pd.merge(left,right,on=["key1","key2"],how="outer"))
```

```python
left = pd.DataFrame({'A': [1, 2], 'B': [2, 2]})
display(left)
right = pd.DataFrame({'A': [1,1,2,4, 5, 6], 'B': [4,3,8,2, 2, 2]})
display(right)
display(pd.merge(left,right,on="A",how="inner"))

display(pd.merge(left,right,on="A",right_index=True,how="inner"))

display(pd.merge(left,right,on="A",how="left"))

display(pd.merge(left,right,on="A",how="right"))
display(pd.merge(left,right,on="A",how="outer"))
display(pd.merge(left,right,on="B"))
```
